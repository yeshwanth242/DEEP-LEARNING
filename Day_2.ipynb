```python
import pandas as pd
import numpy as np

df = pd.read_csv("/content/HR_comma_sep.csv")
df
```

```python
df1 = ['sales','salary']
df_f = pd.get_dummies(df, columns = df1,drop_first=True)
df_f
```

```python
from sklearn.model_selection import train_test_split
tt = train_test_split
x = df_f.drop(['left'],axis=1).values
y = df_f['left'].values
x
```

```python
y
```

```python
x_train,x_test,y_train,y_test = tt(x,y,test_size=0.3)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)
```

```python
from keras.models import Sequential
from keras.layers import Dense

model = Sequential()
model.add(Dense(9,kernel_initializer = 'uniform',activation = 'relu',input_dim = 18))
model.add(Dense(6,kernel_initializer = 'uniform',activation = 'relu',input_dim = 18))
model.add(Dense(1,kernel_initializer = 'uniform',activation = 'sigmoid'))
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.fit(x_train,y_train,epochs=10)
```

```python
y_pred = model.predict(x_test)
y_pred = y_pred.round()
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
accuracy_score(y_test,y_pred)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
df = df_c.drop(['RowNumber','CustomerId','Surname'],axis=1)
df
```

```python
df.shape
```

```python
#One hot encoding of categorical data
df1 =['Geography','Gender']
df_final = pd.get_dummies(df,columns=df1,drop_first=True)
df_final.head()
```

```python
# Split the independent and dependent variables
X = df_final.drop(['Exited'],axis=1)
y = df_final['Exited']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3)
```

```python
# Standardising the data
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
ann = Sequential()

# Input Layer
ann.add(Dense(12, input_dim=11,kernel_initializer='uniform',activation='relu'))

# Hidden layer
ann.add(Dense(8,activation='relu'))

# Output layer
ann.add(Dense(1,activation='sigmoid'))
```

```python
# compile the model
ann.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])

# Fit the model
res = ann.fit(X_train,y_train,batch_size=32,epochs= 120)
```

```python
y_pred = np.round(ann.predict(X_test)).astype(int)
```

```python
cm = confusion_matrix(y_test,y_pred)
cm
```

```python
round(accuracy_score(y_test,y_pred),4)*100
```
